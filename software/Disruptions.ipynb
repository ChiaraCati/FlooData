```python
import pandas as pd
```

```python
#variables 
cod_prov = {'piacenza':'PC', 'ferrara':'FE', 'ravenna':'RA', 'parma':'PR', 'bologna':'BO', 'rimini':'RN', "forli-cesena":'FC', "reggio-nellemilia":'RE', 'modena':'MO'}
```

```python
#upload rendis
rendis = pd.read_csv('/Users/chiara/Documents/GitHub/Owater/datasets/rendis/r_instabilities.csv')
rendis
```

```python
#change rendis missing values into none 
none_rendis = rendis.replace('missing', None)
none_rendis
```

```python
#subset emilia-romagna's instabilities
er_rendis = none_rendis[none_rendis.lot_region == 'emilia-romagna']
er_rendis
```

```python
#drop rendis useless columns 
er_rendis_drop = er_rendis.reset_index(drop=True)
er_rendis_drop=er_rendis_drop.drop(['subtype','lot_location','contract_url','lot_region', 'lot_location_label', 'subtypeLabel', 'contract_lot' ], axis=1)
er_rendis_drop
```

```python
#add date and risk columns 
er_rendis_added_values = er_rendis_drop
er_rendis_added_values['Date'] = None
er_rendis_added_values['Risk'] = None
er_rendis_added_values
```

```python
#change provinces into code 
for idx, row in er_rendis_added_values.iterrows():
    province = row.lot_prov	
    cod = cod_prov[province]
    er_rendis_added_values.loc[idx,'lot_prov'] = cod

er_rendis_added_values
```

```python
#change coilumns names 
er_rendis_added_values.rename(columns = {'instabilities':'Instability', 'type':'Type', 'lot_prov':'Prov_code', 'lot_lat':'Lat', 'lot_long':'Long',}, inplace = True)
rendis_instab = er_rendis_added_values
rendis_instab
```
