```python
#imports
import os
import pandas as pd
```

```python
#open finaldatasets
dir_path = '../datasets/final_dataset'
for item in os.listdir(dir_path):
    print(item)
```

```python
# municipalities with hydrogeological risk
municipality = pd.read_csv(dir_path + '/municipality.csv')
municipality = municipality.drop(municipality.columns[0], axis=1)
municipality
```

```python
# provinces with hydrogeological risk
province = pd.read_csv(dir_path + '/province.csv')
province = province.drop(province.columns[0], axis=1)
province
```

```python
# disruptions translation 
disruptions = pd.read_csv(dir_path + '/disruptions.csv')
disruptions = disruptions.drop(disruptions.columns[0], axis=1)
disruptions = disruptions.replace('Frana', 'Landslide')
disruptions = disruptions.replace('Dissesto Costiero', 'Coastal Disruption')
disruptions = disruptions.replace('fuoriuscite di gas da pozzi per acqua e/o dal terreno', 'Gas leaks from wells for water and / or from the ground'
)
disruptions = disruptions.fillna('Unknown')
disruptions = disruptions.replace('pozzi con presenza di acqua calda', 'Wells with hot water')
disruptions = disruptions.replace('presenza di fratture e/o di sprofondamenti nel terreno', 'Presence of fractures and / or sinking into the ground')
disruptions = disruptions.replace('vulcanetti di fango e/o sabbia', 'Volcanoes of mud and / or sand')

set(disruptions['type'])
disruptions[disruptions.type == 'Unknown']
```

```python
# disruptions cleaning 
set(disruptions['Risk'])
disruptions = disruptions.drop(columns='Risk')
disruptions
```
