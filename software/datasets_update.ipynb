```python
# actions translation 
actions = pd.read_csv(dir_path + '/MD1_CSV.csv')
actions = actions.drop(actions.columns[0], axis=1)


actions.action_description = actions.action_description.fillna('Sconosciuto')
actions.action_description = actions.action_description.replace('Missing','Mancante')

actions_it = set(actions.action_description)

translator = Translator()
source_lan = "it"
translated_to= "en" 
for word in actions_it:
    trans = translator.translate(word, src=source_lan, dest = translated_to)
    actions = actions.replace(word, trans.text)
actions_eng = set(actions.action_description)
actions_eng

```
